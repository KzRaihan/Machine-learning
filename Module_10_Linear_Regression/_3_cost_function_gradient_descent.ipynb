### 1.  How to Calculate Best-fit line of Linear Regression Model

- There are two ways
    - 1. Closed form solution

    - 2. Non-closed from Solution


- 1. Closed Form Solution:
    - Use direct formula

    - find mean square error

- 2. Non-Closed form solution
    - Use Gradient Desent 

```python
# import libaries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step_01: Create the Example Dataset

```python
# Create a dataset

hours = np.array([2,4,6,8,10]).reshape(-1,1)

marks = np.array([55, 60, 70, 80, 88])

hours , marks
```

### Step_02: Fit a linear Regression Model

```python
# import libaries

from sklearn.linear_model import LinearRegression, SGDRegressor
```

```python
# Fit a linear Regression Model (Closed-form solver)

lin_model = LinearRegression()

lin_model.fit(hours, marks)
```

```python
# To see the Learned Parameters (m, c) <-- Data Propertices of LinearRegression Model

print("M: ", lin_model.coef_[0])

print("C: ", lin_model.intercept_)
```

### Stpe_04: Prediction of the model

```python
# prediction of model train data

pred_marks = lin_model.predict(hours)

print("Model Predict Marks: ", pred_marks)
```

```python
# To see the Actual Marks
print("Actual Makrs: ", marks)
```

### Step: 5: Visualizing the Best fit line

```python
# best fit line

# actual data
plt.scatter(hours, marks, color='blue', label="Actual data")

# model predict data

plt.plot(hours, pred_marks, color ='red', label = 'best fit line')

plt.title("Actual data vs Predict data")
plt.xlabel("Hours")
plt.ylabel("Marks")


plt.legend()
plt.grid()

plt.show()
```

### Step_06: Compute the Cost or Loss Function (Mean Square Error)

- Formula:
    - mse = sum(yi - y^i)^2/n

```python
# import libaries
from sklearn.metrics import mean_squared_error
```

```python
# Compute Loss Function or Total error

mse_lin = mean_squared_error(marks, pred_marks)

print("MSE for Linear Regression: ", mse_lin)
```

### Step_07: Use a Gradient Desent Based Model - SGDRegressor

- Gradient Desent best for Higher Dimension and non-linear data points

```python
# import libaries

from sklearn.linear_model import SGDRegressor
```

```python
# Use a Gradient Descent Based model - SGDRegressor

# create an object or SGDRegressor class
sgd_reg = SGDRegressor(
    max_iter = 1000,
    learning_rate = "invscaling",
    eta0 = 0.01,
    random_state = 42
)
```

```python
# fit for SGDRegressor
sgd_reg.fit(hours, marks)

print("M: ", sgd_reg.coef_[0])
print("C: ", sgd_reg.intercept_[0])
```

```python
# predict for SGDRegressor

pred_sgd = sgd_reg.predict(hours)

print("SGD for Prediction: ", pred_sgd)
```

```python
# Actual data points
print("Actual Data points: ", marks)
```

### Step_08: Loss Function or Total Error for Gredient Desent

```python
mse_sgd = mean_squared_error(marks, pred_sgd)

print("MSE of SGDRegressor: ", mse_sgd)
```

```python
# plot for SGDRegressor

# actual data
plt.scatter(hours, marks, color='blue', label="Actual data")

# SGDmodel predict data

plt.plot(hours, pred_sgd, color ='red', label = 'SGD best fit line ')

plt.title("Actual data vs SGD Predict data")
plt.xlabel("Hours")
plt.ylabel("Marks")


plt.legend()
plt.grid()

plt.show()
```

### Step_09: Compare Both lines visually

- The line from LinearRegression (exact solution)

- The line from SGDRegressor (gradient descent solution)

```python
# plot for LinearRegression vs SGDRegressor

# actual data
plt.scatter(hours, marks, color='blue', label="Actual data")

# LinearRegression predict data

plt.plot(hours, pred_marks, color ='red', label = 'LinearRegression best fit line', linestyle="--")

# SGDmodel predict data

plt.plot(hours, pred_sgd, color ='green', label = 'SGD best fit line ', linestyle="-.")

plt.title("LinearRegression vs SGDRegression data")
plt.xlabel("Hours")
plt.ylabel("Marks")


plt.legend()
plt.grid()

plt.show()
```

- for above plot

- LinearRegression best fit line is good compare to SGDRegressor

The key idea:

- **Cost function (MSE)** tells us how good or bad a line is  
- **Gradient descent** (inside `SGDRegressor`) is one way to search for parameters that **reduce this cost**
