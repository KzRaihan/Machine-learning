## Topic: Linear Regression for Build small dataset

### OUTCOMES

- 1. Create a simple dataset

- 2. Train a Linear Regression Model

- 3. View the learned Parameters

- 4. Predict (Validation for existing created data)

- 5. Visualize the Actual data vs Best Fit line

- 6. Prediction for a New data

```python
# import libaries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

### 1. Create a simple dataset

```python
# create dataset

hours = np.array([2, 3, 4, 5, 6, 7, 8, 10]).reshape(-1,1)  # column = 1 but rows = automatically set

marks = np.array([50, 55, 60, 65, 70, 75, 80, 90])

display(hours)

display(marks)
```

### 2. Train a Linear Regression Model

```python
# import Linear Regression model(algorithm)
from sklearn.linear_model import LinearRegression

# Create an object
model = LinearRegression() 

# fit the model

model.fit(hours,marks) 

# here model learn -> m(slope) and c(intercept)
```

###  3. View the learned Parameters

```python
# Learned Parameter (m,c)

m = model.coef_[0]

c = model.intercept_

print("m(slope): ", m)
print("c(Intercept): ", c)
```

- here the best fit line equation:
- y = mx + c

- y = 4.99 * x + 40.00

- for new x the y value is: 
    - y = 4.99 * x + 40.00

### 4. Predict (Validation for existing created data)

```python
predict_marks = model.predict(hours)

print(f"Predicted Marks : {predict_marks}")
```

```python
print(f"Actural Marks: ", marks)
```

### 5. Visualize the Actual data vs Best Fit line

```python
# Actual Data point

# plt.scatter(x = hours, y = marks, color = 'blue', label = 'Actual Data')
plt.scatter(hours, marks, color = 'blue', label = 'Actual Data')


plt.title("Actual Data points")
plt.xlabel("Hours")
plt.ylabel("Marks")

plt.legend()
plt.grid()
```

```python
# Best fit line or predict points

# plt.scatter(x = hours,
#             y = predict_marks,
#             color = 'red',
#             label = "Best fit line")
plt.plot(hours,predict_marks, color='red', label="Best Fit Line")

plt.title("Best fit line or Predictions point")

plt.xlabel("Hours")

plt.ylabel("Marks")

plt.legend()
plt.grid()

plt.show()
```

```python
# Actual data vs Best Fit line

#Visualize the Best Fit Line
plt.scatter(hours, marks, color='blue', label="Actual Data")

plt.plot(hours,predict_marks, color='red', label="Best Fit Line")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks for Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```

### 6. Prediction for a New data

```python
new_hours = np.array([[8.50]])

prediction_marks = model.predict(new_hours)

print(f"for new hours {new_hours} the prediction marks: ", prediction_marks)
```
