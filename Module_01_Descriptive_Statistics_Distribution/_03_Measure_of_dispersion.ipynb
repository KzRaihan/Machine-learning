### 1. Introduction of Measure of Dispersion.

- Definition:
    - Measure of Dispersion describes how spread out or scattered the data values are in a dataset.

- Why need?
    - It provides information about how the data is distributed around the central tendency(mean, median, mode) of dataset.

- common measures of dispersion
    - 1. Variance
    - 2. Standard Deviation
    - so on (Range, IQR)

### 2. Variance and when it's appropriate

- Definition:
    - The variance measures how far(average) each data points is spread out from the mean.

- Formula:
    - Population Variance(σ2) = (∑(xi ​− xˉ)2​)/n 

        - here, xi => each data value
        - xˉ  => mean of the dataset
        - n => numers of data points.

    - Sample Variance (s2) = (∑(xi ​− xˉ)2​)/ n-1.

    - why n - 1?
        - for Bessel's Correction

    - Why square(2)?
        - only find the value not direction(-ve).

    - units of the value is square


```python
# example of variance

data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Score": [70, 72, 68, 90, 60]
}

df = pd.DataFrame(data)

df
```

```python
'''  
Apply formula:
 - Variance(σ2):

      (∑(xi -​ xˉ)2​)
     _______________ 
         n - ddof

- here, 
    - ddof => Delta degrees of freedom.
    - when ddof = 0 (population variance)
    - ddof = 1 [sample variance]

'''
```

```python
# calculate variance using pandas

# population variance
variance_pop = df['Score'].var(ddof = 0)

# sample variance [By deafult]

var_sample = df['Score'].var(ddof = 1)


print("Population Variance: ", variance_pop)
print("Sample Variance: ", var_sample)

```

```python
# find the mean value 

mean_value = df['Score'].mean()

print(f"Mean Value: {mean_value:.2f}")
```

- What is the meaning?
    - Each data point average distance from the mean is 97.6 [in the form of square.]

    - The average difference between score 70 and mean is 97.6.

    - Score 72 average distance from the mean is 97.6.

### 3. Standard Deviation and when it's appropriate.

- Definition:
    - STD measures how much the values deviate (spread out) form the mean on average.
    
    - squre root of variance

    - unit -> as the same of data unit.

```python
# Example of STD

# dataset
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Score": [70, 72, 68, 90, 60]
}

df = pd.DataFrame(data)

df
```

```python
# STD using pandas

std_pop = df['Score'].std(ddof = 0)

std_sam = df['Score'].std(ddof = 1)

print(f"Population Standard Deviation: {std_pop:.2f}")
print(f"Sample Standard Deviation: {std_sam:.2f}")

```

```python
mean_val = df['Score'].mean()
print("Mean value : ", mean_val)
```

- what is meaning of STD 9.88?
    - on average student score deviates 9.88 marks from them mean.

    - or the dataset spread
        - Higest value: (mean + STD)
            - 72.0 + 9.88 => 81.88
        - lowest value: (mean - STD)
            - 72.0 - 9.88 => 62.12 

    - data spread range (62.12 - 81.88)

### 5. Bias-variance Trade-off.

- 1. Overfitting:
    - A high-variance model memorizes the training set.

- 2. Underfit:
    - if Variance is low then model can't proper learns from training data.
