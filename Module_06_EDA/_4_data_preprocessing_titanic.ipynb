### 0. Why Preprocessing is needed

- "Clean data is half the model." 

- Clean data is the foundation of any successful machine learning project.

- Data Preprocessing act as a cleaning the data.


### 1. Handle Missing Values

- Two ways to handle missing value

- 1. Handlde case one: drop
    - if a column has 70% missing value then we drop the column
 
- 2. Handle Missing value case two: fill
    - for numerical features => (median, mean)

    - for categorica features => (mode(most frequenc))

        

### 2. Encoding the Categorical Features

- Definition:
    - Encoding is use for convert categorical data to numerical data.


- Type of Encoding on Categorical Columns
    
    - 1. Label Encoding
        - Best for binary or ordinal categories.
        - Binary: Male = 0, Female = 1
        - Ordinal: Low = 1, Medium = 2, High = 3


    - 2. One-hot Encoding

        - Best for nominal categories with no ranking.

        - when categorical feature values is not binary then we use one-hot encoding.

        - Data are forms in a specific range (0 to 1)

        - Dhaka → [1, 0, 0]

        - Chittagong → [0, 1, 0]


### 3. Normalization and scaling the numerical Features.

- Definition of Scaling
    - Scaling give each feature an equal opportunity to contribute to the learning model.

    - Scaling value can be negative(-0.1, , -1.2).

-  Types of Scaling

    - 1. Z-score Scaling.

    - 2. Min-max Scalling. 

    - 3. Robust scaling.

- Steps of scaling:
    - before split we divide the data into two portation
        - i. features (x)

        - ii. target (y)

    - before scaling we split the data set into two part

        - i. Train
            - train data set need to scaling.

        - ii. Test 
            - test dataset can't need to scaling.

    - Scaling process:
        - StandardScaler (Z-score Normalization).
        - MinMaxScaler (Normalization)
        - RobustScaler scaling.

- Remember: 
    - Scaling doesn't change data relationships—it just helps the model see them clearly, undistracted by large numbers.

    - test data can't be need to scaling.
