### 3. Types of Statistics

- 1. Descriptive Statistics

- 2. Inferiential Statistics

#### 1. Descriptive Statistics

- Descriptive statistics deal with the collection of data, organization, analysis, interpretation and presentation of data.


- NOTE
    - Descriptive or Summarize data.
    - Study of the given (past) data.
    - without prediction.

#### 2. Inferiential Statistics

- Deals with making conclusion and prediction about the Population based on sample.

- Population:   
    - refers to the entire group of individuals or objects that we are interested in studying.

    - eg: if we find the average salary of bangladesh people
    - here, 
        - object of interest => Bangladesh People Salary
        - so, this entire Bangladesh people Salary is Population.


- Sample:
    - Sample is a subset of Population.

    - eg: 
        - Every district some of people's salary is sample.


- NNB:
    - Draw the conclusion about population from a sample.
    - use for prediction.

### 4. Types of Data

- Data is the term used to refer to a set of raw information gathered for a specific purpose.

- when we talk to data then it's form is Tabular(rows, columns)

- Types of Data
    - 1. Categorical or Qualitative data.
    - 2. Numerical or Quantitative data.

#### 1. Categorical or Qualitative data

- Categorical data is the data grouped in the form of categories.

- Types of Categorical Datat
    - 1. Nominal Data.
        - Name only — no order 
        - eg. 
            - Gender group (Male, Female)
            - colors group (red, green, blue)
            - states name, Nationality, Religion
            - Device type, blood group.

    - 2. Ordinal Data. 
        - Order maintain (exists)
        - eg.
            - Education level (Primary, Secondary, Bachelor, Master, PhD)

            - Class Position(1st, 2nd, 3rd, 4th)

            - Size (Small, Medium, Large, Extra Large)

            - Movie Ratings (1 star, 2 stars, 3 stars, 4 stars, 5 stars)

#### 2. Numerical or Quantitative data.

- represent measurable quantities

- Types of Numerical data
    - 1. Discrete Data(integer)
        - represent countable quantities
        - eg:
            - Numbers of students (1,2,3,4,5,6)
            - age (20, 23, 24, 54)
            - rank (1, 2, 3)
    
    - 2. Continuous Data(float)
        - represent measurable quantities
        - eg:
            - Height(5.7, 6.4, 3.8, 4.5) Feet
            
            - weight (63.5, 69.55, 39.99, 45.23) kg
            
            - Temperature (25.4°C, 30.6°C)
            
            - Time (1.5 hours, 2.75 seconds)
            
            - Speed (60.2 km/h, 75.8 km/h)

- NNB:
    - Discrete: countable, integer values
    - Continuous: Measurable, float values

### 5. Why need Statistics for ML

- A Model learns patterns from data.

- Statistic hepls ML model to understand the data before we train models.

- so, if we don't understand data's behavior, then we risk building biased or misleading models.

- Use statistics in ML to study the center, spread, shape, and relationships within data.

- It forms the foundation for preprocessing, features scalling, and model evalution.
