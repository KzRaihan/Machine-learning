### 1. Introduction of ordinal Encoding.

- Definition:
    - Encoding is use for convert categorical data to numerical data.

- Type of Categorical Data
    - 1. nominal data
        - no order data
        - eg: (color, city, gender)
   
    - 2. ordinal data
        - order data
        - eg: (rank, course feedback, review)


- Ordinal Encoding use for ordinal categorical data.


- steps of ordinal encoding:

    - step_01: detect the ordinal column

    - step_02: decide the order
        - for decide order is very important
        - order always maintain sequence or hierarchy
        - order sequence (small < medium < large)


```python
import numpy as np

import pandas as pd
```

### 2. Code Implementation of ordinal encoding.

```python
df = pd.DataFrame({
    "id":[1,2,3,4],
    "color":["red","blue","green","red"],
    "size":["Small","Medium","Large","Medium"],
    "price":[10,12,15,11]
})

df
```

- here, our target column => size
- because size contain the ordinal categorical featues 

```python
# To see how many unique values are present
df['size'].unique()
```

```python
# Step_01: decide the order for each unique column

order = {
    'Small': 1,
    'Medium': 2,
    'Large' : 3
}
```

```python
# Step_03: apply ordinal encoding or convert wholes featues

# way_1: convert and add new column

df['size_encode'] = df['size'].map(order).astype(int)

df
```

```python
# romove the old column

df.drop('size', axis = 1)
```

### ordinal Encoding for real world dataset

```python
df = pd.read_csv("Data.csv")

df
```

- here our target column => Purchased
- because Purchased is a ordinal categorical column

```python
# To see Number of unique categorical 

df['Purchased'].unique()
```

- so, here two unique values
    - No
    - Yes

```python
# step_01: decide the order
# No = 0
# Yes = 1

order = {
    'No': 0,
    'Yes': 1
}
```

```python
# step_02: apply ordinal encoding

df['Purchased'] = df['Purchased'].map(order).astype(int)

df
```

- so, Purchased colums are ready for model learning.
