### 1. What is EDA?

- Definition:
    - Initial Exploration of Raw data that called Exploratory Data Analysis (EDA).

    - To Understand the Raw data.


- Position of EDA
    - In ML Pipeline, EDA exists in between Data Collection and Data Preprocessing.

    - EDA is a continuous processes.

- Diagram of EDA

    - Raw Data----> EDA ------> Conclusion of Raw data

### 2. Why Need EDA?

- Need:
    - To Building ML Model

    - Analysis and reporting.

    - Handling Missing Values.

    - Detecting Outliers.

    - To helps Features Engineering.

    - Validate Assumpations.

### 3. Outcomes of EDA

- Outcomes:

    - Understanding data distributions, correlations, and patterns.

    - Visualizing data with histograms, scatter plots, heatmaps, and pairplots.

    - Identifying data imbalance and relationships between variables(columns).

    - EDA best practices before model building  

### 4. Steps of EDA

- Step_01: Labels or Categories the features/columns

    - 1. Numerical Column

    - 2. Categorical Column

    - 3. Target Column


- Step_02: Univariate Analysis 

    - Analysis Each features 

    - Types of Univariate Analysis:
    
        - 1. univariate analysis

        - 2. bivariate analysis

        - 3. Multivariate analysis

- Step_03: Conclusion or Summarizes

    - Summarize the findings of the EDA and make decisions about how to proceed with further analysis.

#### Step_01: In Details (Labels or Categories the features/columsn)

- 1. Numerical Column:

    - If Columns contain continuous data, then it's a numerical column
        - eg: Age, RestingBP, PassengerId
        - Format: (10, 10.5, 20, 60, 12, 45, 34, 89)

    
- 2. Categorical Column

    - Normal case: 
        - If columns contains categorical data, then it's consider as categorical column
        
        - eg: name, gender/sex, ChestPainType, Ticket etc

        - Format: (Yes, No) or (F, M) or (T123, U900, 123)

    - Expection case:
        - If a columsn contains numerical data but actually act as categorical such as

        - eg: passengerâ€™s ticket class (1st, 2nd, 3rd), number of siblings/spouses aboard (1,2,0, 3,5) 

        - if a column data withn (1-10) then it's consider as categorical column when dataset are larger (1000 features) [-Fix_Number_of_range(1,10)]


- 3. Target Column
    - Model Class column
    - eg: Survived(0,1) or (No, Yes), (class1, class2, class3)

#### in details Step_02: Univariate Analysis 

- 1. Univariate analysis:

    - Analysis the single column individually.

- Steps of Univariate Analysis:

    - 1. Distribution analysis of numerical columns

        - Using Descriptive Statistics:
            - i. To identify the shape
                - Normal, Skewed, uniform distribution

            - ii. Central Tendency of data
                - mean, median, mode

            - iii. Disperation(spread) of data
                - variance, standard derivation, IQR(Q3-Q1)

    - 2. Visualizations
        - By using different plots or graphs to show the data distribution

        - Type of plot for numerical column:
            - histogram, scatterplot, boxplot etc.

        - Provide:
            - Identify the skewness, outliers and missing values

            - Skewness:
                - To show the data distribution.

                - i). Right/Positive Skew
                    - If Mixmum data in left side or 2-d plot.
                    - When skewed value positive (4.23, 2,33, 1,33)

                - ii). Left/Negative Skew
                    - If Mixmum data in Right side or 2-d plot.
                    - When Skewed value is negative (-2.33, -0.52, -3.22)

                - iii). Symmetic Skew 
                    - if plot look like bell Curve (https://www.investopedia.com/terms/b/bell-curve.asp)
                    - When skewed value is zero or near to zero (0.22, 0, 0.11, 0. 21)

            - Outliers Detection:
                - Extreme value is called outliers
                - To decide whether to include(if important data) or exclude(if not important data) them from the analysis


            - To helps to identify Missing Value

                - Decide how to handle them

                - Handlde case one: drop
                    - if a column has 70% missing values then we drop the column

                - Handle Missing value case two: fill
                    - fill with (median, mean, mode)
   


- 2. bivariate analysis

- After complete Univariate analysis then we move or foucus on bivariate analysis to find or extract more information from Raw datasets.

- Types of Bivariate Analysis

    - 1. Numerical to Numerical.

    - 2. Numerical to Categorical.

    - 3. Categorical to Categorical.
